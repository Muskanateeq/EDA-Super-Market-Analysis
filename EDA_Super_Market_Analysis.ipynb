# The "Date" column was previously in object format, which represents strings, and it has now been converted into the correct date format.

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

# The "Time" column has also been converted from object format to the correct time format.

```python
df['Time'].unique()
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M', errors='coerce').dt.time
df['Time'] = pd.to_datetime(df['Time'], errors='coerce').dt.time
df.info()

```

# The value count has been calculated to check the dataset by category, in order to identify which branch has the highest value count.

# A pie chart of the Branch category has been created to visually represent the total occurrences, making it easier to understand.

```python
# prompt: Plot pie graph for distribution between types in data

import matplotlib.pyplot as plt

# Assuming 'Branch' is the column you want to plot
type_counts = df['Branch'].value_counts()

plt.figure(figsize=(5, 3))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Branches')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

# The branches have been used to visualize the gender distribution, allowing us to determine which branch has more females and which branch has more males. The data has been grouped into two categories for this purpose.

```python
# Group by Branch and Gender, then count the occurrences
branch_gender_counts = df.groupby(['Branch', 'Gender']).size()

# Display the result
print(branch_gender_counts)
```

# INSIGHT # 1 I plotted the counts of gender across different branches to visualize the distribution and compare the number of males and females in each branch. This analysis helps to easily identify which branches have a higher representation of either gender.

```python
# prompt: plot graph for branch and its gender

import matplotlib.pyplot as plt

# Group by Branch and Gender, then count the occurrences
branch_gender_counts = df.groupby(['Branch', 'Gender']).size().unstack()

# Create the bar plot
branch_gender_counts.plot(kind='bar', figsize=(8, 6))
plt.title('Gender Distribution by Branch')
plt.xlabel('Branch')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.legend(title='Gender')
plt.show()
```

# INSIGHT # 2 Created a bar graph to visually represent the counts of each product line, which helps to identify which products are selling the most. This visualization allows for a straightforward comparison between different product lines, highlighting the top-selling items within the dataset.

```python
# prompt: plot bar graph for Product_line with color palete

import matplotlib.pyplot as plt

# Get the count of each Product_line
type_counts = df['Product_line'].value_counts()

# Create the bar plot with a color palette
plt.figure(figsize=(10, 6))
type_counts.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen', 'orange', 'violet', 'pink'])
plt.title('Distribution of Product Line')
plt.xlabel('Product Line')
plt.ylabel('Number of Products')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

# INSIGHT # 3 Which branch has which main product.This visualization helps to identify which products are selling the most of each branch.

```python
# insight-3 Which branch has which main product

import matplotlib.pyplot as plt
# Group by Branch and Product Line, then count the occurrences
branch_product_counts = df.groupby(['Branch', 'Product_line']).size().unstack()

# Create the bar plot
branch_product_counts.plot(kind='bar', figsize=(17, 10))
plt.title('Product Line Distribution by Branch')
plt.xlabel('Branch')
plt.ylabel('Number of Products')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.legend(title='Product Line')
plt.show()
```

```python
# prompt: plot box plot and distribution productline

import matplotlib.pyplot as plt
import seaborn as sns

# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Product_line', y='Unit_price', data=df)
plt.title('Box Plot of Unit Price by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Unit Price')
plt.xticks(rotation=45, ha='right')
plt.show()

# Distribution plot (using a violin plot for better visualization)
plt.figure(figsize=(10, 6))
sns.violinplot(x='Product_line', y='Unit_price', data=df)
plt.title('Distribution of Unit Price by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Unit Price')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_columns = ['Unit_price', 'Quantity', 'Tax_5', 'Total', 'cogs', 'gross_income', 'Rating']

for column in numerical_columns:
  plt.figure(figsize=(10, 6))
  plt.subplot(1, 2, 1)
  sns.boxplot(x=df[column])
  plt.title(f'Box Plot of {column}')

  plt.subplot(1, 2, 2)
  sns.histplot(df[column], kde=True)
  plt.title(f'Distribution of {column}')

  plt.tight_layout()
  plt.show()
```

# Created a correlation graph for all numerical features in the dataset to visualize the relationships and associations between them. This graph helps identify the strength and direction of correlations, allowing us to understand how different numerical variables relate to each other. By analyzing these correlations, we can uncover potential patterns and insights that can inform further analysis or decision-making.

```python
# prompt: plot correlation graph for all numerical features

import matplotlib.pyplot as plt
import numpy as np

# Select only numerical features for correlation analysis
numerical_features = df.select_dtypes(include=np.number)

# Calculate the correlation matrix
correlation_matrix = numerical_features.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

# Plotted a bivariate graph for all features in the dataset to visualize the relationships between them. This analysis helps to identify potential correlations and interactions between different variables, providing insights into how they influence each other. By examining these relationships, we can better understand the underlying patterns in the data.

```python
# prompt: plot bivariate graph for all feature among one another

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical features for bivariate analysis
numerical_features = df.select_dtypes(include=np.number)

# Create a pair plot to visualize relationships between all numerical features
sns.pairplot(numerical_features)
plt.show()
```
